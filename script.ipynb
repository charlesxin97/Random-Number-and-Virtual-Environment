# Step1: Create random_set_1 <1000 random number valued between 1 and 100>

```python
import random
import json
import matplotlib.pyplot as plt
sample_a = []
for i in range(1000):
    sample_a.append(random.randint(0,100))
json_a = json.dumps(sample_a)

w = open('output/random_set_1.json', 'w')
w.write(json_a)
w.close()

x = range(1000)
y = sample_a
plt.scatter(x, y, s = 0.3)
plt.show()
```

# Step2: Create random_set_2 based on dataset_a using equation (y=3x+6)

```python
with open("output/random_set_1.json",'r') as load_f:
    sample_a=json.load(load_f)

sample_b = []
for num in sample_a:
    sample_b.append(num*3+6)
json_b = json.dumps(sample_b)

w = open('output/random_set_2.json', 'w')
w.write(json_b)
w.close()

x = range(1000)
y = sample_b
plt.scatter(x, y, s = 0.3)
plt.show()
```

# Step3: Visualization of random_set_1 and random_set_2, 1 as x_axis and 2 as y_axis

```python
with open("output/random_set_1.json",'r') as load_f:
    sample_a=json.load(load_f)
with open("output/random_set_2.json",'r') as load_f:
    sample_b=json.load(load_f)

plt.plot(sample_a, sample_b, color='blue')
plt.title("visualization")
plt.xlabel("index")
plt.ylabel("value")
plt.savefig("output/c.png")
plt.show()
```
